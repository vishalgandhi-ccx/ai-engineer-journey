```python
import numpy as np

py_list = [1, 2, 3, 4, 5]
np_arr  = np.array([1, 2, 3, 4, 5])


print(f"array: {np_arr}, type: {type(np_arr).__name__}, dtype: {np_arr.dtype}")

import time

N = 1_000_000
py_list = list(range(N))
np_arr  = np.arange(N)

print(f"list: {py_list}, type: {type(py_list).__name__}")

# Python way
t0 = time.perf_counter()
py_result = [x * 2 for x in py_list]
py_time = time.perf_counter() - t0

# NumPy way
t0 = time.perf_counter()
np_result = np_arr * 2
np_time = time.perf_counter() - t0


print(f"Python list: {py_time*1000:.2f} ms")
print(f"NumPy array: {np_time*1000:.2f} ms")
print(f"Speedup:     {py_time/np_time:.1f}x")
```
